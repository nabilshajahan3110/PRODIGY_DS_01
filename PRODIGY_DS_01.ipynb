## TASK 01 - Create Bar Plot/Histogram

Create a bar plot/histogram to visualize distribution of categorical or continuous variables.

### Loading Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

### Loading Dataset

```python
# Load the dataset
data=pd.read_csv("Renatl Bike Sharing system hr.csv")
data
```

### Data Exploration

### Bar Plot

```python
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
data['season_label'] = data['season'].map(season_map)

sns.countplot(x='season_label', data=data)
plt.title('Distribution of Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

### Histogram

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['cnt'], bins=30, kde=True)
plt.title('Distribution of Bike Rental Counts')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.show()
```
